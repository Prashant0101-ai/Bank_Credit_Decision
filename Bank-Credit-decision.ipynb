# Multivariate Analysis for crucial variables

```python
plt.figure(figsize=[11,7])

plt.title("Heatmap for Client's Family and Income\n", fontdict={'fontsize':21,'fontweight':6,'color':'Grey'})
sns.heatmap(inp3[['NAME_CONTRACT_TYPE_x','NAME_CASH_LOAN_PURPOSE','NAME_CONTRACT_STATUS','NAME_PAYMENT_TYPE','NAME_CLIENT_TYPE','CODE_GENDER','FLAG_OWN_CAR','FLAG_OWN_REALTY','CNT_CHILDREN','AMT_INCOME_TOTAL','NAME_INCOME_TYPE','NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS','NAME_HOUSING_TYPE','DAYS_EMPLOYED','CNT_FAM_MEMBERS']].corr(),annot=True,cmap="Reds")
plt.show()
```

## Conclusion
1) Under loan type, "consumer loan" turns out to be highest in demand and "revolving loan" to be lowest in demand. From target variable we can calculate the percentage of beong a defaulter.Under loan type we can see that bank must offer minimum to "revolving loan" and highest to "consumer loan".

2) Bank must offer higher loans to females than males as latter has higher percentage of being default and further demand for loan is higher for females.

3) Bank must reduce its offer to "Maternity leave" as they have higher risk of defaulting whereas "Working" income type have highest demand for loan and repayment rate is good too.

4) Bank must offer loan to clients having residing in a Office apartment or own House and discourage from clients living in rented apartment or with parents. However demand for loans is highest for clients living in own House.

5) Clients having higher number of children, Bank must not offer them a loan as their percentage of being default is high.

6) There is a negative correlation between number of childern and total income of client.Client having less number of children are good to offer loan by banks.

7) Bank must offer loan to clients having highest education as academic degree insted of lower secondary.

8) Bank must offer loan to Clients who are married as they have lower percentage of being default whereas single have highest percentage of being default.

## Hence, above conclusion can guide banks to most important decision of making loan offer to most credible clients.
